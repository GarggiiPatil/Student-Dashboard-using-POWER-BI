```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#DATA GENERATION
np.random.seed(42)
num_students=500

departments = ['CSE', 'ECE', 'ME', 'CE', 'IT', 'EE']
semesters = ['Sem1', 'Sem2', 'Sem3', 'Sem4', 'Sem5', 'Sem6', 'Sem7', 'Sem8']
subjects = ['Maths', 'Physics', 'Chemistry', 'CS', 'Electronics', 'Mechanics']
grades = ['A', 'B', 'C', 'D', 'F']
genders = ['Male', 'Female']


data = {
    'StudentID': [f"S{1+1:03d}" for i in range(num_students)],
    'Name': [f"Student_{i+1}" for i in range(num_students)],
    'Gender': np.random.choice(genders, num_students),
    'Department': np.random.choice(departments, num_students),
    'Semester': np.random.choice(semesters, num_students),
    'Subject': np.random.choice(subjects, num_students),
    'Marks': np.random.randint(30, 100, num_students),
    'Attendance (%)': np.random.randint(50, 100, num_students),
    'Grade': np.random.choice(grades, num_students),
    'Date': pd.date_range(start='2024-07-01', periods=num_students, freq='D')
}

df = pd.DataFrame(data)
```

```python
#BASIC ANALYSIS
print("Null Values:\n", df.isnull().sum())
print("\nAverage marks per department:\n", df.groupby("Department")["Marks"].mean())
```

```python
#VISUALIZATION
plt.figure(figsize=(10, 5))
sns.barplot(x='Department', y='Marks', data=df)
plt.title('Average Marks per Department')
plt.tight_layout()
plt.show()

# Attendance trend over time
plt.figure(figsize=(10, 5))
df.set_index('Date')['Attendance (%)'].resample('W').mean().plot()
plt.title('Weekly Attendance Trend')
plt.ylabel('Attendance (%)')
plt.tight_layout()
plt.show()
```

```python
#Export to CSV for POWER BI
output_path = 'student_performance_dataset.csv'
df.to_csv(output_path, index=False)
print(f"\nDataset Exported to: {output_path}")
```
